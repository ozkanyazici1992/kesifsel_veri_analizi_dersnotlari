!!!! keşifsel veri analizi (kısaca EDA/Exploratory data analysis) !!!!

Bize bir veri seti verildiğinde, nihai hedefimiz şöyle bir şey olur:

- gelecek ayın konut fiyatlarının tahmini.
- önümüzdeki haftadaki çalkantılı müşterileri tahmin etmek.
- bir bitkinin yarınki elektrik tüketimini ve enerji talebini tahmin etmek.
- bir fotoğraftan kişiyi tespit edebilen bir mobil uygulama oluşturmak.
- sürekli video yayını yapan bir web sitesi için öneri sistemi uygulamak. vb.

Verilerin kalitesi hakkında konuştuğumuzda, genellikle veri temizleme'yi kastederiz. Veri temizleme, verileri veya üzerlerinde oluşturduğumuz modelleri yanlış anlamamıza yol açabilecek alakasız veya gürültülü değerleri eleme işlemini ifade eder.

Verilerimizi temizledikten sonra, hedefimizi açıklayabilecek faydalı özellikleri bulmamız gerekir. Hedef, veri kümemizdeki açıklamaya çalıştığımız değişkeni ifade eder, özellik ise hedefi açıklamak için kullanılan değişkenleri belirtir. Faydalı özellikler bulma ve yaratma sürecine özellik mühendisliği denir.

Ancak, iyi özellikler bulmak için verilerimizi analiz etmemiz gerekir. Bunu, özellikler arasındaki ve özellikler ile hedef arasındaki ilişkileri analiz ederek yapıyoruz. Bu sürece veri keşfi denir.

Keşifsel veri analizi kısaca veri temizliği, veri keşfi ve özellik mühendisliği aşamalarının birleşimidir.

# 1. Keşifsel veri analizi nedir?
Keşifsel veri analizi, veri bilimi boru hattının ilk yapı taşıdır. Modelleme aşamasına geçmeden önce, verilerimizdeki ilişkileri anlamamız gerekir. Verilerimizi ne kadar iyi anlarsak, modellerimizin en iyi performanslarını elde etmelerine yardımcı olacak en kullanışlı özellikleri daha başarılı bir şekilde belirleriz.

EDA'yı üç aşamadan veya adımdan oluşan yinelemeli bir süreç olarak kavramsallaştırırız. Aşağıdaki şekil bu aşamaları ve yinelemenin nasıl gerçekleştiğini göstermektedir:

1. Veri temizleme: EDA sürecinin ilk aşamasıdır. Veri kümesini alırız ve verilerde tespit edebileceğimiz sorunları gidermeye çalışırız.

2. Veriyi keşfetme: Bu ikinci adım, farklı özellikler ile verilerdeki modeller arasındaki ilişkileri keşfetmek için verilerin araştırılmasıdır. Bu aşamada kullandığımız temel araçlar, istatistikler ve görselleştirme teknikleridir. Araştırma sırasında verilerdeki yeni sorunlar tespit edersek, temizleme aşamasına geri döner ve yeni bulunan sorunları ortadan kaldırarak verileri temizleriz.

3. Özellik mühendisliği: EDA'nın son adımı, en kullanışlı özellikleri seçtiğimiz veya mevcut özelliklerden yeni özellikler oluşturduğumuz özellik mühendisliğidir. Bu aşama bizi veri bilimi boru hattının sonraki adımlarına götürecektir. Ancak verilerle ilgili bazı sorunlar tespit edersek, veri temizleme aşamasına başvurmalı ve verileri tekrar temizlemeliyiz. Ayrıca, mevcut özellikleri dönüştürürsek veya yenilerini yaratırsak, bu yeni özelliklerin kullanışlı olup olmadığını test etmek için veri keşfi aşamasına geri dönmemiz gerekir.
